```python
from google.colab import files
uploaded = files.upload()  # Upload your flag image
```

```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.preprocessing import image

# Load the uploaded flag image
img_path = list(uploaded.keys())[0]  # Get the uploaded image filename
img = image.load_img(img_path, target_size=(224, 224))  # Resize image for the model

# Preprocess the image
img_array = image.img_to_array(img)  # Convert to array
img_array = np.expand_dims(img_array, axis=0)  # Add batch dimension
img_array /= 255.0  # Normalize the image

# Display the image
plt.imshow(img)
plt.axis('off')  # Remove axis
plt.show()
```

```python
# Load the MobileNetV2 model pre-trained on ImageNet
model = tf.keras.applications.MobileNetV2(weights='imagenet')

# Make a prediction
predictions = model.predict(img_array)

# Decode the predictions (ImageNet class labels)
decoded_predictions = tf.keras.applications.mobilenet_v2.decode_predictions(predictions, top=5)[0]

# Print top 5 predictions
for i, (imagenet_id, label, score) in enumerate(decoded_predictions):
    print(f"{i + 1}: {label} ({score:.2f})")
```
